Klay Thompson didn't play the 2020 season, due to injury and for the rest of the players, 2020 was their rookie season

So we collect a third dataset of aggregate player statistics
- taken from https://www.kaggle.com/datasets/drgilermo/nba-players-stats?select=player_data.csv

```python
#Player data
dfplayerdata = pd.read_csv("data/player_seasonal/all_seasons.csv")
dfplayerdata.head(-1)
```

And when we encounter missing seasonal values, we can either find their last played season (in case of injury) or some cursory information on a rookie player, such as age, height, weight. If we want to, we can also get some draft information on the player, from the 'dfplayerdata' dataframe

```python
ps_pergame = []
for p in game["PLAYER_NAME"]:
    p_pergame = s20_pergame[s20_pergame["player"] == p]
    p_pergame = p_pergame.drop(["season","lg"],axis=1)
    p_pergame.rename(columns={"player":"PLAYER_NAME"},inplace=True)
    p_bio = dfplayerdata.loc[dfplayerdata["player_name"] == p] #get player bio
    p_bio = p_bio.loc[p_bio["season"] == p_bio["season"].max()]
    if p_pergame.empty: #Player didn't play last season
        p_pergame = p_pergame.append(pd.Series(0, index=p_pergame.columns), ignore_index=True)
        p_pergame["PLAYER_NAME"] = p     
    if len(p_pergame) > 1: #player switched teams last season
        current_team = game.loc[game["PLAYER_NAME"] == p]["TEAM_ABBREVIATION"].iloc[0]
        if not p_pergame.loc[p_pergame["tm"] == current_team].loc[p_pergame["g"] < 10].empty: #Player played less than 10 games in new team
            p_pergame = p_pergame.loc[p_pergame["tm"] == "TOT"] #Select totals
        else:
            p_pergame = p_pergame.loc[p_pergame["tm"] == current_team] #Select current team performance        
    p_pergame["weight"] = p_bio["player_weight"].iloc[0]
    p_pergame["height"] = p_bio["player_height"].iloc[0]
    ps_pergame.append(p_pergame)
ps_pergame = pd.concat(ps_pergame, axis=0)
```

```python
dfplayerpergame.rename(columns={"player":"PLAYER_NAME"},inplace=True)
dfplayerdata.rename(columns={"player_name":"PLAYER_NAME"},inplace=True)

season = 2021
ps_pergame = []
games = dfgames[dfgames["SEASON"] == season]
game = games.iloc[0]
game = dfgamebox[dfgamebox["GAME_ID"] == game["GAME_ID"]]
last_season_pergame = dfplayerpergame[dfplayerpergame["season"] == season-1]
for p in game["PLAYER_NAME"]:
    p_pergame = last_season_pergame[last_season_pergame["PLAYER_NAME"] == p]
    p_bio = dfplayerdata.loc[dfplayerdata["PLAYER_NAME"] == p] #get player biometric info
    p_bio = p_bio.loc[p_bio["season"] == p_bio["season"].max()]
    if p_pergame.empty: #Player didn't play last season
        p_pergame = dfplayerpergame[(dfplayerpergame["PLAYER_NAME"] == p) & (dfplayerpergame["season"] < season)] #has the player played previously (out due to injury)
        if not p_pergame.empty:
            p_pergame = p_pergame.loc[p_pergame["season"] == p_pergame["season"].max()] #Take last season played
        else: #first season
            p_pergame = p_pergame.append(pd.Series(0, index=p_pergame.columns), ignore_index=True)
            p_pergame["PLAYER_NAME"] = p
            p_pergame["age"] = 20 #Age set at 20 for newly drafted player - could be generative
    if len(p_pergame) > 1: #player switched teams last season
        current_team = game.loc[game["PLAYER_NAME"] == p]["TEAM_ABBREVIATION"].iloc[0]
        if not p_pergame.loc[p_pergame["tm"] == current_team].loc[p_pergame["g"] < 10].empty: #Player played less than 10 games in new team
            p_pergame = p_pergame.loc[p_pergame["tm"] == "TOT"] #Select totals
        else:
            p_pergame = p_pergame.loc[p_pergame["tm"] == current_team] #Select current team performance        
    ps_pergame.append(p_pergame)
ps_pergame = pd.concat(ps_pergame, axis=0)
ps_pergame = ps_pergame.drop(["season","lg"],axis=1)
ps_pergame
```
